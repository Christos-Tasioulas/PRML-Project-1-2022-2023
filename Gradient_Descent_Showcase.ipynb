## Ερώτημα 2: Υπολογισμός παραγώγων

### Υποερώτημα 2.1: Χειρόγραφες πράξεις

Έστω $f(x) = x^TAx + b^Tx$, μια συνάρτηση d μεταβλητών, με $A$ συμμετρικό πίνακα (δηλ.
$A = A^T$) διάστασης d x d, και $b ∈ ℝ^d$. Να υπολογίσετε αναλυτικά βήμα προς βήμα με το χέρι,
την παράγωγο της $f(x)$ ως προς το διάνυσμα x ∈ ℝ^d παραθέτωντας όλα τα ενδιάμεσα βήματα.

Θα χρησιμοποιήσουμε από το Matrix Cookbook (https://www.math.uwaterloo.ca/~hwolkowi/matrixcookbook.pdf)
τις σχέσεις $(33)-(45)$ για να υπολογίσουμε την παράγωγο της $f(x)$.

Έχουμε:

$$
f(x) = x^TAx + b^Tx \quad(1)
$$ 
και 
$$
\nabla f(x) = \begin{pmatrix}
\frac{d f(x)}{dx_1} \\
\vdots \\
\frac{d f(x)}{dx_d} \\
\end{pmatrix}
\quad(2)
$$

Παραγωγίζουμε την $(1)$: 
$$
d f(x) = d (x^TAx + b^Tx)\quad(3)
$$

Χρησιμοποιούμε την σχέση $(35)$ από το Matrix Cookbook:
$$
d f(x) = d (x^TAx) + d (b^Tx)\quad(4)
$$

Χρησιμοποιούμε την σχέση $(34)$ από το Matrix Cookbook:
$$
d f(x) = Ad (x^Tx) + b^Td (x)\quad(5)
$$

Χρησιμοποιούμε την σχέση $(37)$ από το Matrix Cookbook:
$$
d f(x) = A[d (x^T)x + x^Td (x)]+ b^Td (x)\quad(5)
$$

Τέλος, χρησιμοποιούμε την σχέση $(44)$ από το Matrix Cookbook και έχουμε:
$$
d f(x) = A[(d x)^Tx + x^Td (x)]+ b^Td (x)\quad(6)
$$

Ενσωματώνουμε την $(6)$ στην $(2)$ και έχουμε:
$$
\nabla f(x) = \begin{pmatrix}
\frac{A[(d x)^Tx + x^Td (x)]+ b^Td}{dx_1} \\
\vdots \\
\frac{A[(d x)^Tx + x^Td (x)]+ b^Td}{dx_d} \\
\end{pmatrix}
\quad(7)
$$

### Yποερώτημα 2.2: Χειρόγραφες πράξεις

Έστω $Α, Β, Χ$ πίνακες πραγματικών αριθμών διάστασης d x d με τάξη (rank) ίση με d. Nα
βρείτε, αναλυτικά και σε κλειστή μορφή (closed form solution) το ολικό ελάχιστο του
παρακάτω προβλήματος βελτιστοποίησης:

$$
\min_{X} \{ \left\lVert A-XB \right\rVert _F^2\}
$$

Να εξηγήσετε με λεπτομέρεια όλη τη διαδικασία σκέψης που θα ακολουθήσετε και να
παραθέσετε όλα τα ενδιάμεσα βήματα των πράξεων.

Έστω: 
$$
f(X) = \left\lVert A-XB \right\rVert _F^2 \quad(8)
$$
Θέλουμε να βρούμε την ελάχιστη τιμή της συνάρτησης ως προς τον πίνακα Χ. Αυτό επιτυγχάνεται όταν $\nabla _X f(X) = 0$. 
Για να βρούμε το $\nabla _X f(x)$ θα χρησιμοποιήσουμε σχέσεις από το προαναφερθεν Matrix Cookbook όπως οι σχέσεις $(33)-(45)$ αλλά και η σχέση $(132)$.

Γνωρίζουμε ότι:

$$
\nabla _Χ f(X) =
= \begin{pmatrix}
\frac{d f(X)}{dx_{11}} & \dots & \frac{d f(X)}{dx_{1d}}\\
\vdots & \ddots & \vdots \\
\frac{d f(X)}{dx_{d1}} & \dots & \frac{d f(X)}{dx_{dd}}\\
\end{pmatrix}
\quad(9)
$$

Για να ισχύει ότι $\nabla _X f(X) = 0$ πρεπει $\forall x_{ij}, 1 \leq i,j \leq d$ να ισχύει ότι $\frac{d f(X)}{dx_{ij}} = 0$.
Άρα πρέπει να βρούμε την $\frac{d f(X)}{dx_{ij}}$.

Για να την βρούμε, αρχικά παραγωγίζουμε την $(8)$:

$$
\frac{d}{dΧ} f(X) = \frac{d}{dΧ} \left\lVert A-XB \right\rVert _F^2 \quad(10)
$$

Θέτω:
$$
Υ = A-XB \quad(11)
$$

Όπου:

$$
ΧΒ = Α-Υ \quad(12)   
$$

Αφού πολλαπλσιάσουμε με $Β^{-1}$ στο τέλος κάθε μέλους θα έχουμε:

$$
ΧΒ*Β^{-1} = (A-Υ)Β^{-1} \Rightarrow Χ = (A-Υ)Β^{-1} \quad(13) 
$$

Παραγωγίζουμε και έχουμε ότι:

$$
dX = -dY*B^{-1} \quad(14)
$$

Ενσωματώνουμε την $(11)$ και την $(14)$ στην $(8)$ (Κανόνας της Αλυσίδας) και έχουμε:

$$
\frac{d}{dΧ} f(X) = \frac{d}{-dΥ*B^{-1}} \left\lVert Υ \right\rVert _F^2 \quad(15)
$$

Πολλαπλασιάουμε τον αριθμητή και τον παρονομαστή του κλάσματος στο 2ο μέλος με Β στο τέλος και έχουμε:

$$
\frac{d}{dΧ} f(X) = \frac{d*Β}{-dΥ*B^{-1}*Β} \left\lVert Υ \right\rVert _F^2 \Rightarrow \frac{d}{dΧ} f(X) = \frac{d}{dΥ} B*\left\lVert Υ \right\rVert _F^2 \quad(16)
$$

Χρησιμοποιούμε την σχέση $(132)$ από το Matrix Cookbook:

$$
\frac{d}{dX} f(X) = \frac{d}{dΥ} [-B*Tr((Υ)(Υ)^H)] \quad(17)
$$

Χρησιμοποιούμε την σχέση $(34)$ από το Matrix Cookbook:

$$
\frac{d}{dX} f(X) = -B* \frac{d}{dΥ} [Tr((Υ)(Υ)^H)] \quad(18)
$$

Χρησιμοποιούμε την σχέση $(132)$ από το Matrix Cookbook:

$$
\frac{d}{dX} f(X) = -Β*2*Υ \quad(19)
$$

Αντικαθιστούμε την $(11)$ στην $(19)$:

$$
\frac{d}{dX} f(X) = -B*2(Α-ΒΧ) = 2Β^2Χ-2BA \quad(20)
$$

Πρέπει να λύσουμε την εξίσωση $\frac{d}{dX} f(X) = 0$.
Έχουμε:

$$
2Β^2Χ-2BA = 0 \Rightarrow 2B^2X = 2BA \Rightarrow B^2X = ΒΑ
$$

Πολλαπλασιάζουμετα δύο μέλη της εξίσωσης με $B^{-2}$ στην αρχή:

$$
B^{-2}*B*A = B^{-2}*B^2*X \Rightarrow X = B^{-1}A \quad(21)
$$

Το ολικό ελάχιστο του προβλήματος ως προς Χ προκύπτει όταν αντικαταστήσουμε το Χ με το $B^{-1}A$.
Έχουμε:

$$
\min_{X} \{ \left\lVert A-XB \right\rVert _F^2\} = \left\lVert A-B^{-1}AB \right\rVert _F^2 \quad(22)
$$

Και επειδή οι πίνακες είναι d*d με τάξη d μπορούμε να χρησιμοποιήσουμε την ταυότητα του Woodbury και να βρούμε ότι:

$$
\min_{X} \{ \left\lVert A-XB \right\rVert _F^2\} = \left\lVert A-AB^{-1}B \right\rVert _F^2 = \left\lVert A-AI \right\rVert _F^2 = 0
$$

## Ερώτημα 3: Gradient Descent

Πρώτα εισάγουμε την matplotlib στον αλγόριθμό μας

```python
import matplotlib.pyplot as plt
```

Έπειτα ορίζουμε τις συνάρτησεις f1, f2 και τις παραγώγους τους ως προς κάθε μεταβλητή (x1, x2)

```python

""" Functions to be used in the gradient descent algorithm """

# f1 given function from the exercise sheet
def f1(x1, x2):
    return (x1-2)*(x1-2)+(x2-3)*(x2-3)

# derivative of f1 function with respect to x1 variable
def df11(x1, x2):
    return 2*(x1-2)

# derivative of f1 function with respect to x2 variable
def df12(x1, x2):
    return 2*(x2-3)

# f2 given function from the exercise sheet
def f2(x1, x2):
    return (1 - (x2-3)*(x2-3)) + 20*((x1+3)-(x2-3)*(x2-3))*((x1+3)-(x2-3)*(x2-3))

# derivative of f2 function with respect to x1 variable
def df21(x1, x2):
    return 40*((x1+3)-(x2-3)*(x2-3))

# derivative of f2 function with respect to x2 variable
def df22(x1, x2):
    return -2*(4-x2) - 80*((x1+3)-(x2-3)*(x2-3))*(x2-3)
```

Εδώ είναι ο Gradient Descent αλγόριθμος που υλοποιήθηκε

Σημείωση: ο αλγόριθμος για τα x1,x2 δεν χρησιμοποιεί prev μεταβλητές για να κρατάει τις τιμές της προηγούμενης επανάληψης και χρησιμοποιεί, για παράδειγμα, την τιμή της x1 στην τωρινή επανάληψη για την εύρεση της x2. Αυτό είναι μια μικρή βελτιστοποίηση για να γίνονται γρηγορότερα οι προσεγγίσεις.

```python

""" Gradient descent function that uses a function with two variables at a constant learning rate """

def gradient_descent(x1, x2, f, df1, df2, T, learning_rate):

    # intitializing the values to be plotted
    f_values = [f(x1, x2)]
    iterations = [0]
    
    # gradient descent algorithm
    for i in range(T):

        # x = prev_x - λ *df(x1, x2) 
        x1=x1-learning_rate*df1(x1, x2)
        x2=x2-learning_rate*df2(x1, x2)

        # finding the current value pf the given function
        cur_result = f(x1, x2)

        # appending the current values in the to be plotted list
        f_values.append(cur_result);
        iterations.append(i+1);

    # Create a plot with a red dashed line
    plt.plot(iterations, f_values)

    # Add labels to the axes
    plt.xlabel('Iterations')
    if f == f1:
        plt.ylabel('f1 Values')
    elif f == f2:
        plt.ylabel('f2 Values')
    else:
        plt.ylabel('F Values')

    # Set the plot title
    plt.title('Gradient Descent')

    # Show the plot
    plt.show()
```

Χρησιμοποιώντας τις παραμέτρους που δόθηκαν στην εκφώνηση (Starting x1: 0  Starting x2: 0  learning rate: 0.5  total iterations: 10), σύμφωνα με τα γραφήματα της matplotlib παρατηρούμε ότι όσο αφορά την f1, ο αλγόριθμος συγκλίνει άμεσα στην 1η επανάληψη, ενώ στην f2, ο αλγόριθμος βγάζει άπειρες τιμές στην 3η επανάληψη.
Άρα καταλαβαίνουμε ότι το 0.5 είναι μεγάλη τιμή για learning rate.

```python
# initialize parameters
learning_rate = 0.5
x1 = 0
x2 = 0
T = 10

print("Starting x1:", x1, " Starting x2:", x2," learning rate:", learning_rate, " total iterations:", T)
gradient_descent(x1, x2, f1, df11, df12, T, learning_rate)
gradient_descent(x1, x2, f2, df21, df22, T, learning_rate)
```

Άμα μεγαλώσουμε κι άλλο το learning rate, γύρω στο 0.75, παρατηρούμε ότι στην f2 "σκάει" επίσης στην 3η επανάληψη, ενώ στην f1 συγκλίνει πιο αργά στην 3η επανάληψη.

```python
print("Starting x1:", x1, " Starting x2:", x2," learning rate:", 0.75, " total iterations:", T)
gradient_descent(x1, x2, f1, df11, df12, T, 0.75)
gradient_descent(x1, x2, f2, df21, df22, T, 0.75)
```

Άμα μειώσουμε το learning rate στο 0.01 παρατηρούμε ότι στην f1 πλησιάζει σε σύγκλιση στη 10η επανάληψη και στην f2 "σκάει" στην 5η επανάληψη.

```python
print("Starting x1:", x1, " Starting x2:", x2," learning rate:", 0.01, " total iterations:", T)
gradient_descent(x1, x2, f1, df11, df12, T, 0.01)
gradient_descent(x1, x2, f2, df21, df22, T, 0.01)
```

Παρατηρούμε ότι με το learning rate στο 0.005, ο αλγόριθμος σταματάει να "σκάει" στην f2 και πλησιάζει σε σύγκλιση στην 4η επανάληψη ενώ στην f1 συνεχίζει να πλησιάζει σε σύγκλιση στην 10η επανάληψη.
